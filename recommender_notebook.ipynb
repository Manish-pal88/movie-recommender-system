```python
movies=pd.read_csv("tmdb_5000_movies.csv")
credits=pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies = movies.merge(credits, on="title")
movies.head(5) # input data is ready.
```

# data cleaning and preprocessing 

```python
movies= movies[['movie_id','genres','overview','keywords','title','cast','crew']]
```

```python
movies.info()
```

```python
# checking for missing value and gives the count of it.
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)# since the missing value are very low we remove that row(movies) of the dataset.
```

```python
# checking for duplicates 
movies.duplicated().sum()# currently here are no duplicate values 
```

```python
# how the iloc function works here
movies.iloc[0].genres
```

```python
import ast# ast was need to import carefully 
def converter(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i["name"])
    return l
```

```python
movies["genres"]=movies["genres"].apply(converter)
```

```python
movies["keywords"]=movies["keywords"].apply(converter)
```

```python
# we have to consider only top 5 cast name to correct it.
import ast
def converter3 (obj):
    l = []
    count=0
    for i in ast.literal_eval(obj):
        if(count!=5):
            l.append(i["name"])
            count+=1
        else:
            break
    return l
```

```python
movies["cast"]=movies["cast"].apply(converter3)
```

```python
movies.head(5)
```

```python
movies.iloc[0].cast
```

```python
# now we are working on the crew column i wants to fetch the director name from the crew colun for that first i wants to crete a function for that
def fetch_director (obj):
    l=[]
    for i in ast.literal_eval(obj):
        if (i['job']=='Director'):
            l.append(i['name'])
            break
    return l   

```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.iloc[0].crew
```

```python
# we are converting overview strings into list so that i can create tags easily.
movies['overview']=movies['overview'].apply(lambda x: x.split())
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(' ','') for i in x ] )
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(' ','') for i in x ] )
movies['cast']=movies['cast'].apply(lambda x:[i.replace(' ','') for i in x ] )
movies['crew']=movies['crew'].apply(lambda x:[i.replace(' ','') for i in x ] )
```

```python
movies.head(2)
```

```python
movies['tags']= movies['overview']+movies['genres']+movies['keywords']+movies['cast']+ movies['crew']
```

```python
movies.head()
```

```python
new_df= movies[['movie_id','title','tags']]
```

```python
new_df
```

```python
new_df = new_df.reset_index(drop=True) # correcting the indexing 
```

```python
new_df
```

```python
new_df['tags']= new_df['tags'].apply(lambda x:" ".join(x))
```

```python
# covert every word into lower case so that will be helpful to us a lot
new_df['tags']=new_df['tags'].apply(lambda x: x.lower())
```

```python
# vectorizer
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=5000,stop_words= 'english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem (text):
    y=[]

    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)   
    
```

```python

new_df['tags'].apply(stem)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

# recommender system

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors) # it is showing the similarity b/w  movies . 
```

```python

new_df
```

# recommender function(it will recommend top 5 movies which are related to input movies )


```python

def recommend(movie):
    movie_index= new_df[new_df['title']==movie].index[0]
    distances= similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)), reverse= True ,key = lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
recommend('Shanghai Calling')
```

```python
manual_test_cases = [
    ('Iron Man', ['Iron Man 2', 'Iron Man 3', 'Avengers']),
    ('Avatar', ['Titan A.E.', 'Independence Day', 'John Carter']),
    ('The Dark Knight Rises', ['The Dark Knight', 'Batman Begins']),
]

manual_precision = []
for movie, expected in manual_test_cases:
    movie_idx = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_idx]
    rec_indices = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:4]
    recommendations = [new_df.iloc[i[0]]['title'] for i in rec_indices]
    
    hits = set(recommendations[:5]) & set(expected)
    precision = len(hits) / 3
    manual_precision.append(precision)
    
    print(f"\n📽️ {movie}")
    print(f"   Got: {recommendations[:3]}")
    print(f"   Expected: {expected[:3]}")
    print(f"   Manual Precision@3: {precision:.0%}")

avg_manual_precision = sum(manual_precision) / len(manual_precision)
print(f"\n AVERAGE MANUAL PRECISION@3: {avg_manual_precision:.0%}")
```
